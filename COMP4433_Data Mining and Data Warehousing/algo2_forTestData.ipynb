```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Visualise all columns in the dataframe
pd.pandas.set_option('display.max_columns', None)

#Read the testing dataset
test_df=pd.read_csv('test.csv')
```

```python
#Handle missing values in categorical features
catFeaturesWithNA=[feature for feature in test_df.columns if test_df[feature].isnull().sum()>1 and test_df[feature].dtypes=='O']
```

```python
def handleCatNAs(dataset,catfeaturesNA):
    data=dataset.copy()
    data[catfeaturesNA]=data[catfeaturesNA].fillna('Missing')
    return data
```

```python
test_df=handleCatNAs(test_df,catFeaturesWithNA)
test_df[catFeaturesWithNA].isnull().sum()
```

```python
#Handle missing values in numerical features
numFeaturesWithNA=[feature for feature in test_df.columns if test_df[feature].isnull().sum()>1 and test_df[feature].dtypes!='O']
```

```python
for feature in numFeaturesWithNA:
    median_value=test_df[feature].median()
    test_df[feature+'NA']=np.where(test_df[feature].isnull(),1,0)
    test_df[feature].fillna(median_value,inplace=True)
```

```python
test_df[numFeaturesWithNA].isnull().sum()
```

```python
#Handle time-related variables
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
       test_df[feature]=test_df['YrSold']-test_df[feature]
```

```python
test_df.head()
```

```python
#Handle skewed numerical values: log normal distribution
import numpy as np
skewed_num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea']

for feature in skewed_num_features:
    test_df[feature]=np.log(test_df[feature])
```

```python
test_df.head()
```

```python
#Handle categorical variables with less than 1% occurrence
categorical_features=[feature for feature in test_df.columns if test_df[feature].dtype=='O']
```

```python
for feature in categorical_features:
    temp=test_df.groupby(feature)['Id'].count()/len(test_df)
    temp_df=temp[temp>0.01].index
    test_df[feature]=np.where(test_df[feature].isin(temp_df),test_df[feature],'RareValue')
```

```python
test_df.head()
```

```python
#Turn each value into its index
for feature in categorical_features:
    labels_ordered=test_df.groupby(feature)['Id'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    test_df[feature]=test_df[feature].map(labels_ordered)
```

```python
test_df.head()
```

```python
featuresExceptIdSaleprice=[feature for feature in test_df.columns if feature not in ['Id','SalePrice'] ]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(test_df[featuresExceptIdSaleprice])
```

```python
scaler.transform(test_df[featuresExceptIdSaleprice])
```

```python
#Transform the training dataset and add back the Id and SalePrice attributes
data = pd.concat([test_df[['Id']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(test_df[featuresExceptIdSaleprice]), columns=featuresExceptIdSaleprice)],
                    axis=1)
```
