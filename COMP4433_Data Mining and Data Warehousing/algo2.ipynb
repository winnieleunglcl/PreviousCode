```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Visualise all columns in the dataframe
pd.pandas.set_option('display.max_columns', None)

#Read the training dataset
df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
#Handle missing values in categorical features
catFeaturesWithNA=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes=='O']
```

```python
def handleCatNAs(dataset,catfeaturesNA):
    data=dataset.copy()
    data[catfeaturesNA]=data[catfeaturesNA].fillna('Missing')
    return data
```

```python
df=handleCatNAs(df,catFeaturesWithNA)
df[catFeaturesWithNA].isnull().sum()
```

```python
#Handle missing values in numerical features
numFeaturesWithNA=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes!='O']
```

```python
for feature in numFeaturesWithNA:
    median_value=df[feature].median()
    df[feature+'NA']=np.where(df[feature].isnull(),1,0)
    df[feature].fillna(median_value,inplace=True)
```

```python
df[numFeaturesWithNA].isnull().sum()
```

```python
#Handle time-related variables
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
       df[feature]=df['YrSold']-df[feature]
```

```python
df.head()
```

```python
#Handle skewed numerical values: log normal distribution
import numpy as np
skewedNumFeatures=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in skewedNumFeatures:
    df[feature]=np.log(df[feature])
```

```python
df.head()
```

```python
#Handle categorical variables with less than 1% occurrence
categorical_features=[feature for feature in df.columns if df[feature].dtype=='O']
```

```python
for feature in categorical_features:
    temp=df.groupby(feature)['SalePrice'].count()/len(df)
    temp_df=temp[temp>0.01].index
    df[feature]=np.where(df[feature].isin(temp_df),df[feature],'RareValue')
```

```python
df.head()
```

```python
#Turn each value into its index
for feature in categorical_features:
    labels_ordered=df.groupby(feature)['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[feature]=df[feature].map(labels_ordered)
```

```python
df.head()
```

```python
featuresExceptIdSaleprice=[feature for feature in df.columns if feature not in ['Id','SalePrice'] ]
```

```python
#Scale & translate each feature such that its value is between zero and one
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(df[featuresExceptIdSaleprice])
```

```python
scaler.transform(df[featuresExceptIdSaleprice])
```

```python
#Transform the training dataset and add back the Id and SalePrice attributes
data = pd.concat([df[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(df[featuresExceptIdSaleprice]), columns=featuresExceptIdSaleprice)],
                    axis=1)
```

```python
data
```

```python
data.to_csv('trainedTrain.csv',index=False)
```

```python
dataset=pd.read_csv("trainedTrain.csv")
```

```python
#Capture the dependent feature
y_train=dataset[['SalePrice']]
```

```python
#Drop dependent features from dataset
X_train=dataset.drop(['Id','SalePrice'],axis=1)
```

```python
#shrink the coefficients for those attributes that do not contribute much to the prediction task - SalePrice
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
featureSelectionModel = SelectFromModel(Lasso(alpha=0.005, random_state=0))
featureSelectionModel.fit(X_train, y_train)
```

```python
featureSelectionModel.get_support()
```

```python
selected_feat = X_train.columns[(featureSelectionModel.get_support())]
print('Total features: {}'.format((X_train.shape[1])))
print('Selected features: {}'.format(len(selected_feat)))
print('Features with coefficients shrank to zero: {}'.format(
    np.sum(featureSelectionModel.estimator_.coef_ == 0)))
```

```python
selected_feat
```

```python
X_train=X_train[selected_feat]
```

```python
X_train.head()
```

```python
import xgboost
regressor=xgboost.XGBRegressor()
```

```python
#Hyper Parameter Optimization
n_estimators = [100, 500, 900, 1100, 1500]
max_depth = [2, 3, 5, 10, 15]
booster=['gbtree','gblinear']
learning_rate=[0.05,0.1,0.15,0.20]
min_child_weight=[1,2,3,4]
base_score=[0.25,0.5,0.75,1]

# Define the grid of hyperparameters to search
hyperparameter_grid = {
    'n_estimators': n_estimators,
    'max_depth':max_depth,
    'learning_rate':learning_rate,
    'min_child_weight':min_child_weight,
    'booster':booster,
    'base_score':base_score
    }
```

```python
#Set up the random search with 4-fold cross validation
import xgboost
from sklearn.model_selection import RandomizedSearchCV
random_cv = RandomizedSearchCV(estimator=regressor,
            param_distributions=hyperparameter_grid,
            cv=5, n_iter=50,
            scoring = 'neg_mean_absolute_error',n_jobs = 4,
            verbose = 5, 
            return_train_score = True,
            random_state=42)
```

```python
random_cv.fit(X_train,y_train)
```

```python
random_cv.best_estimator_
```

```python
regressor=xgboost.XGBRegressor(base_score=0.25, booster='gbtree', colsample_bylevel=1,
             colsample_bynode=1, colsample_bytree=1, enable_categorical=False,
             gamma=0, gpu_id=-1, importance_type=None,
             interaction_constraints='', learning_rate=0.05, max_delta_step=0,
             max_depth=2, min_child_weight=4, 
             monotone_constraints='()', n_estimators=900, n_jobs=4,
             num_parallel_tree=1, predictor='auto', random_state=0, reg_alpha=0,
             reg_lambda=1, scale_pos_weight=1, subsample=1, tree_method='exact',
             validate_parameters=1, verbosity=None)
```

```python
regressor.fit(X_train,y_train)
```

```python
import pickle
filename = 'finalized_model.pkl'
pickle.dump(regressor, open(filename, 'wb'))
```
